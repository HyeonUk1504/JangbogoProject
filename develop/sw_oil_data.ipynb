```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_excel('국제_석유제품가격20100101-20211013.xls')
df=df.drop(index=[0,1,2,3,4,5], axis=0)
df=df.rename(columns={'Unnamed: 1':'경유(0.001%)','Unnamed: 2':'경유(0.05%)'})
df=df.reset_index(level=None,
                      drop=False, 
                      inplace=False, 
                      col_level=0,
                       col_fill='')
df
```

```python
#df.shape
#df.info()
#pd.to_datetime(df['국제유가석유제품'])

df['국제유가석유제품']=df['국제유가석유제품'].str[:-1]
```

```python
df['국제유가석유제품']=df['국제유가석유제품'].str.replace('년','-')
```

```python
df['날짜']=df['국제유가석유제품'].str.replace('월','-')
```

```python
df['날짜']=pd.to_datetime(df['날짜'])
```

```python
df
```

```python
#df.info()

plt.plot(df['날짜'], df['경유(0.05%)'])
```
